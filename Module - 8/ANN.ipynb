```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\batch -3\Machine Learning\Module - 8\PotatoPrice.csv")
df
```

```python
plt.xlabel('Potato in kilogram(kg)')
plt.ylabel('price in Taka')
plt.scatter(df.potato_kg, df.price)
```

```python
x = df[['potato_kg']]
print(x)
```

```python
y = df[['price']]
print(y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=1)
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

ANN Model Building

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(1,input_shape=[1],activation='linear')])
```

Compile The ANN model

```python
model.compile(optimizer='sgd',loss='mean_squared_error',metrics=['accuracy'])
```

Train ANN Model

```python
model.fit(xtrain, ytrain, epochs=500)
```

Predict

```python
model.predict(xtest)
```

```python
ytest
```

Evaluate The Model Loss and Accuracy

```python
test_loss,test_accc=model.evaluate(xtest,ytest,verbose=2)
```

Predict The Output Manually

```python
import numpy as np

# Give any unknown potato kilogram value,to know the price
#(N.B: the potato kilogram value have to be any value upto 1,for the decent prediction. Since our fitted data potato_kg range is 1 to 7)

input_data = np.array([[1000]])  # Add an extra dimension to match 2D input
print(model.predict(input_data))
```
