```python
import pandas as pd 
import pandas as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(r"D:\batch -3\Machine Learning\Module - 8\PotatoPrice.csv")
df
```

Visualization data in graphical representation(Not necessary, but helps to understand the dataset)

```python
plt.xlabel('potato in kg')
plt.ylabel('price in tk')
plt.scatter(df.potato_kg,df.price)
```

```python
x = df[['potato_kg']]

y = df[['price']]
```

Split Dataset for Train and Test

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=1)

xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ytest
```

Train dataset using model/algorithm

```python
# use the regression model for the dataset
reg = LinearRegression()
reg.fit(xtrain,ytrain)
```

```python
reg.predict(xtest)
```

```python
ytest
```

Accuracy of The Model

```python
reg.score(xtest,ytest)
```

predict the output manually
