```python
import pandas as pd

data = {'Category': ['A','B','C','A', 'B', 'C', 'A', 'B', 'C']}

df = pd.DataFrame(data)
df
```

```python
one_hot_encoded_df = pd.get_dummies(df, columns=['Category'])
one_hot_encoded_df
```

```python
one_hot_encoded_df = pd.get_dummies(df,columns=['Category'],prefix ='Dumy')
one_hot_encoded_df
```
